```python
import os
```

```python
path = "waste_processed"

for root, dirs, files in os.walk(path):
    folder_parts = root.split(os.sep)
    print("FOLDER PARTS:", folder_parts)
    print("count files:", len(files))
    
```

```python
target_path = "waste_balanced"
os.makedirs(target_path, exist_ok=True)

for root, dirs, files in os.walk(path):
    rel_path = os.path.relpath(root, path)
    
    if rel_path == "." :
        dest_root = target_path 
    else:
        dest_root = os.path.join(target_path, rel_path)
    os.makedirs(dest_root, exist_ok=True)

    for d in dirs:
        os.makedirs(os.path.join(dest_root, d), exist_ok=True)

print(f'Folder structure copied from "{path}" to "{target_path}".')
```

```python
import random
import shutil

k = 944
image_exts = {".jpg"}

for root, dirs, files in os.walk(path):
	rel_path = os.path.relpath(root, path)
	if rel_path == "." :
		dest_root = target_path 
	else:
		dest_root = os.path.join(target_path, rel_path)
	os.makedirs(dest_root, exist_ok=True)

	image_files = [f for f in files if os.path.splitext(f)[1].lower() in image_exts]
	if not image_files:
		continue

	selected = random.sample(image_files, min(k, len(image_files)))

	for fname in selected:
		src_file = os.path.join(root, fname)
		dst_file = os.path.join(dest_root, fname)
		shutil.copy2(src_file, dst_file)

	print(f"{rel_path}: copied {len(selected)} images")
```

```python
source_base = "waste_processed"
target_base = target_path  # already "waste_balanced"

# find the "glass" folder in source (keeps relative structure)
plastic_rel_paths = []
for r, ds, fs in os.walk(source_base):
    if os.path.basename(r).lower() == "plastic":
        plastic_rel_paths.append(os.path.relpath(r, source_base))

if not plastic_rel_paths:
    raise FileNotFoundError('No "plastic" folder found under waste_processed.')

rel_plastic = plastic_rel_paths[0]
src_plastic = os.path.join(source_base, rel_plastic)
dst_plastic = os.path.join(target_base, rel_plastic)
os.makedirs(dst_plastic, exist_ok=True)

src_images = [f for f in os.listdir(src_plastic) if os.path.splitext(f)[1].lower() in image_exts]
dst_images = {f for f in os.listdir(dst_plastic) if os.path.splitext(f)[1].lower() in image_exts}

candidates = [f for f in src_images if f not in dst_images]
n_to_add = min(2, len(candidates))

if n_to_add < 2:
    print(f'Only {n_to_add} new images available to add (requested 31).')

to_copy = random.sample(candidates, n_to_add)

for fname in to_copy:
    shutil.copy2(os.path.join(src_plastic, fname), os.path.join(dst_plastic, fname))

print(f'Added {n_to_add} images to "{dst_plastic}". New total: {len(dst_images) + n_to_add}')
```
